```python
def clean_Price_per_sqft_data(data_set):
    dataOut = pd.DataFrame()
    for k,s in data_set.groupby('Location'):
        mn = np.mean(s['Price/Sqft'])
        std = np.std(s['Price/Sqft'])
        new_data_set = s[(s['Price/Sqft']>(mn-std)) & (s['Price/Sqft']<=(mn+std))]
        dataOut = pd.concat([dataOut,new_data_set],ignore_index=True)
    return dataOut

df_new = clean_Price_per_sqft_data(df)
df_new.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python

def scatterPlot(data_set,Location):
    bhk2 = data_set[(data_set.Location == Location) & (data_set.BHK==2)]
    bhk3 = data_set[(data_set.Location==Location) & (data_set.BHK==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.Sqft,bhk2.Price,color='blue', label ='2 BHK', s=50)
    plt.scatter(bhk3.Sqft,bhk3.Price,marker='+',color='green', label='3 BHK', s=50)
    plt.xlabel("Total Square Feet area")
    plt.ylabel("Price Per Square Feet")
    plt.title(Location)
    plt.legend()
    
scatterPlot(df_new,"Keshtopur")
```
