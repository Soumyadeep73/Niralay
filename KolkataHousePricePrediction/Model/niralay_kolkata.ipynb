```python
def clean_Price_per_sqft_data(data_set):
    dataOut = pd.DataFrame()
    for k,s in data_set.groupby('Location'):
        mn = np.mean(s.price_per_sqft)
        std = np.std(s.price_per_sqft)
        new_data_set = s[(s.price_per_sqft>(mn-std)) & (s.price_per_sqft<=(mn+std))]
        dataOut = pd.concat([dataOut,new_data_set],ignore_index=True)
    return dataOut

data_set5 = clean_Price_per_sqft_data(data_set4)
data_set5.shape
```

```python
data_set5.head()
```

```python
def scatterPlot(data_set,location):
    BHK2 = data_set[(data_set.Location == location) & (data_set.BHK==2)]
    BHK3 = data_set[(data_set.Location==location) & (data_set.BHK==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(BHK2.Sqft,BHK2.Price,color='blue', label ='2 BHK', s=50)
    plt.scatter(BHK3.Sqft,BHK3.Price,marker='+',color='green', label='3 BHK', s=50)
    plt.xlabel("Total Square Feet area")
    plt.ylabel("Price Per Square Feet")
    plt.title(location)
    plt.legend()
    
scatterPlot(data_set5,"Alipore")
```

```python
def clean_bhk_data(data_set):
    exclude_indices = np.array([])
    for l, LDataSet in data_set.groupby('Location'):
        bhkStats = {}
        for bhk, bhkDataSet in LDataSet.groupby('BHK'):
            bhkStats[bhk] = {
            'mean' : np.mean(bhkDataSet.Sqft),
            'std' : np.std(bhkDataSet.Sqft),
            'count' : bhkDataSet.shape[0]
        }
        for bhk,bhkDataSet in LDataSet.groupby('BHK'):
            stats = bhkStats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhkDataSet[bhkDataSet.Sqft<(stats['mean'])].index.values)
    return data_set.drop(exclude_indices,axis='index')


data_set6 = clean_bhk_data(data_set5)
data_set6.shape
```

```python
x = data_set9.drop('Price', axis='columns')
x.head()
```

```python
y = data_set9['Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
#from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos={
        'linear_regression' : {
            'model' : LinearRegression(),
            'params': {
                'normalize':[True, False]
            }
        },
        'lasso':{
            'model':Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random', 'cyclic']
            }
        },
        'decision_tree':{
            'model':DecisionTreeRegressor(),
            'params':{
                'criterion':['mse','friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv= ShuffleSplit(n_splits =5, test_size = 0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'],config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model':algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
        
        
    return pd.DataFrame(scores,columns = ['model', 'best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

```python
def predict_price(location,Sqft,BHK):
    loc_index = np.where(x.columns == location)[0][0]
    
    X = np.zeros(len(x.columns))
    X[0] = Sqft
    X[1] = BHK
    if loc_index >= 0:
        X[loc_index] = 1

    return lr_clf.predict([X])[0]
```
